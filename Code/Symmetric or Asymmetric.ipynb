```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data
data = np.random.normal(loc=0, scale=1, size=2000)

# Create histogram
plt.hist(data, bins=30, color='skyblue', alpha=0.7)

# Define μ and σ
mu = 0  # mean
sigma = 1  # standard deviation

# Define all x-tick points
x_ticks = [
    mu - 3*sigma, mu - 2*sigma, mu - sigma,
    mu, mu + sigma, mu + 2*sigma, mu + 3*sigma
]

# Define cutoff points for dashed lines (only specific points)
cutoff_points = [mu - 3*sigma, mu - sigma, mu + sigma, mu + 3*sigma]

# Add red dashed lines at specified cutoff points
for cutoff in cutoff_points:
    plt.axvline(cutoff, color='red', linestyle='dashed', linewidth=1)

# Customize x-ticks with labels
x_labels = [
    r'$\mu-3\sigma$', r'$\mu-2\sigma$', r'$\mu-\sigma$', r'$\mu$',
    r'$\mu+\sigma$', r'$\mu+2\sigma$', r'$\mu+3\sigma$'
]
plt.xticks(x_ticks, x_labels)

# Customize labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Data with Symmetry')

# Display plot
plt.show()

```
